# Graph analysis

```python
import numpy as np
from bct import distance, clustering
from numpy.polynomial.polynomial import polyfit
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
import seaborn as sns
import functions as st
from scipy import stats
from tqdm import tqdm
from scipy.stats.stats import pearsonr

parc = 'Sch240'
mode = 0
behaviour_list = ['APM','NART_IQ','Q1_TotalWords','Q6_TotalCorrect','CoC_abs_rev']
_, CM= st.load_data(parcellation=parc,
                         n_connectomes='20',
                         behaviour_list=behaviour_list, load_nifti=False)

# load mode weights
cca_mat = np.loadtxt('cca_weights_mat_mode'+str(mode)+'.csv',delimiter=',')
```

## Simple nodal strengths & clustering coefficient

```python
# structural connectivity = group average
SC = np.mean(CM['NoLesion'],axis=2)

# node strength
node_strength = np.mean(SC+SC.T,axis=1)

# clustering coefficient
CC = clustering.clustering_coef_wu(SC+SC.T)

# cca node strength
# I've chosen strength rather than degree as this will account for the number of connections a give node might have
# Also note that it is an absolute value
cca_node_strength = np.mean(abs(cca_mat)+abs(cca_mat.T),axis=1)
cca_node_strength_signed = np.mean(cca_mat+cca_mat.T,axis=1)
```

## Loss of global efficiency (vulnerability)

```python
# the whole graph efficiency without any node lesions
tot_eff = distance.efficiency_wei(SC+SC.T)
eff = []

for node in tqdm(range(240)):
    SCtmp = SC+SC.T
    
    # lesion the node
    SCtmp[node,:] = 0
    SCtmp[:,node] = 0
    
    # recalculate efficiency
    eff.append(tot_eff - distance.efficiency_wei(SCtmp))
eff = np.array(eff)
```

# Brain plots

```python
# Retrieve destrieux parcellation in fsaverage5 space from nilearn
from nilearn import plotting,datasets
import nibabel as nib
import matplotlib.pyplot as plt

def surface_plot(data,vmax,cmap='viridis',symmetric_cbar=False,high_res=False,title=None):
    # get sch atlas
    if high_res is True:
        path = '/Users/luke/Documents/Projects/StrokeNet/Docs/Atlas/Schaefer200/Parcellations/FreeSurfer5.3/fsaverage/label/'
    else:
        path = '/Users/luke/Documents/Projects/StrokeNet/Docs/Atlas/Schaefer200/Parcellations/FreeSurfer5.3/fsaverage5/label/'

    sch_atlas_l = nib.freesurfer.io.read_annot(path+'lh.Schaefer2018_200Parcels_7Networks_order.annot')
    sch_atlas_r = nib.freesurfer.io.read_annot(path+'rh.Schaefer2018_200Parcels_7Networks_order.annot')
    parcellation_l = sch_atlas_l[0]
    parcellation_r = sch_atlas_r[0]

    # generate roi values
    cca_values_l = np.zeros(np.shape(parcellation_l))
    cca_values_r = np.zeros(np.shape(parcellation_r))

    for parcel in range(100):
        cca_values_l[parcellation_l==(parcel+1)] = data[parcel]
        cca_values_r[parcellation_r==(parcel+1)] = data[parcel+100] #add 100 for right hand side

    # Retrieve fsaverage5 surface
    if high_res is True:
        fsaverage = datasets.fetch_surf_fsaverage('fsaverage')
    else:
        fsaverage = datasets.fetch_surf_fsaverage()

    # plot the surface (left)
    f, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'projection': '3d'})
    f.set_figheight(4)
    f.set_figwidth(12)
    
    plotting.plot_surf_stat_map(fsaverage['pial_left'], stat_map=cca_values_l, 
                                hemi='left',
                                title='Title',
                                cmap=cmap,
                                colorbar=False,
                                symmetric_cbar=symmetric_cbar,
                                threshold=0,
                                bg_on_data=True,
                                vmax = vmax ,
                                bg_map=fsaverage['sulc_left'],
                                alpha=1,axes=ax1,
                               darkness=0.6)

    # plot the surface (right)
    plotting.plot_surf_stat_map(fsaverage['pial_right'], stat_map=cca_values_r, 
                                hemi='right',
                                title='Title',
                                cmap=cmap,
                                colorbar=False,
                                symmetric_cbar=symmetric_cbar,
                                threshold=None,
                                bg_on_data=True,
                                vmax = vmax ,
                                bg_map=fsaverage['sulc_right'],
                                alpha=1,axes=ax2,
                               darkness=0.6)
    if title is not None:
        plt.savefig(title,dpi=600)
    plotting.show()
    
def subcortex_plot(data,vmax,cmap='viridis',symmetric_cbar=False,slices=[-42,0],title=None):
    # load the roi-parcellation
    parcellation_file = '/Users/luke/Documents/Projects/StrokeNet/Docs/Atlas/Schaefer200/Schaefer200_plus_HOAAL.nii'
    n_nodes = 240
    img_parc = nib.load(parcellation_file)
    img_parc_data = img_parc.get_data()

    # transfer data values to nifti array
    img_array = np.zeros(np.shape(img_parc_data))
    n_rois = np.int(np.max(img_parc_data))
    for roi in range(n_rois):
        if roi < 200:
            img_array[img_parc_data==roi+1] = 0
        else:
            img_array[img_parc_data==roi+1] = data[roi]

    # create img object
    img = nib.Nifti1Image(img_array, img_parc.affine)
    plotting.plot_stat_map(img,
                           threshold=0,
                           colorbar=True,
                           cmap=cmap,
                           display_mode='z',
                           vmax = vmax,
                           symmetric_cbar=symmetric_cbar,
                           cut_coords=(slices))
    if title is not None:
        plt.savefig(title,dpi=600)
    plotting.show()
```

```python
surface_plot(node_strength,vmax=np.max(abs(node_strength)),high_res=False)
subcortex_plot(node_strength,vmax=np.max(abs(node_strength)))

surface_plot(eff,vmax=np.max(abs(eff)),high_res=False)
subcortex_plot(eff,vmax=np.max(abs(eff)))

surface_plot(cca_node_strength_signed,vmax=0.25,cmap='bwr',symmetric_cbar=True,high_res=True,title='/Users/luke/Documents/Projects/StrokeNet/Docs/Figures/2020/CCA_nodal_surface.jpeg')
subcortex_plot(cca_node_strength_signed,vmax=0.25,cmap='bwr',symmetric_cbar=True,title='/Users/luke/Documents/Projects/StrokeNet/Docs/Figures/2020/CCA_nodal_subcortex.svg')
```

# Scatter plots

```python
plt.figure(figsize=(4,2))

# strength plot
plt.subplot(1,2,1)

#get x and y
x = cca_node_strength
y = node_strength

# report correlation
r,p = pearsonr(x,y)
print('r=',r,', p=',p)

# Fit trend line with polyfit
b, m = polyfit(x,y, 1)

# plot
plt.scatter(x, y, s=10, linewidth=0,color=[0.5,0.5,0.5],alpha=0.5)
plt.plot(x, b + m * x, '-', linewidth=1, color='k')

plt.xlabel("CCA loading strength",fontsize=12)
plt.ylabel("Node strength",fontsize=12)
st.plotting_tools.remove_border()

# GE plot
plt.subplot(1,2,2)
x = cca_node_strength
y = eff

# report correlation
r,p = pearsonr(x,y)
print('r=',r,', p=',p)

# Fit trend line with polyfit
b, m = polyfit(x,y, 1)

# plot
plt.scatter(x, y, s=10, linewidth=0,color=[0.5,0.5,0.5],alpha=0.5)
plt.plot(x, b + m * x, '-', linewidth=1, color='k')

plt.xlabel("CCA loading strength",fontsize=12)
plt.ylabel("Node vulnerability",fontsize=12)
st.plotting_tools.remove_border()
plt.tight_layout()
plt.savefig('/Users/luke/Documents/Projects/StrokeNet/Docs/Figures/2020/graph_theory_scatter.svg',dpi=600)
plt.show()
```
